##PART 1 : WINE QUALITY

the red wine dataset, publicly available since 2009 and detailed in [Cortez et al., 2009], offers valuable insights into the physicochemical properties and sensory quality of Portuguese "Vinho Verde" red wine. With 1599 instances and 11 input attributes, the dataset presents an opportunity for classification or regression tasks. Given the unbalanced class distribution and potential attribute correlations, careful consideration of outlier detection and feature selection methods is advisable for robust analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

#1st--->>: Loading Wine Quality Dataset
rwd =pd.read_csv('/content/drive/MyDrive/wineQualityReds.csv')
print(rwd)

#2nd---> :
print("Wine Quality Dataset - ")

print(rwd.head())

#3rd--->>>> :
print("\nNumber of Records available in Wine Quality Datset -  ", len(rwd))

#4th--->>>>>:
print("\nNumber of Columns and their Description - Wine Quality Datset :")
print(rwd.info())

#5th--->>>>:
rwd_cleaned = rwd.dropna()

print("Dataset - after Deleting NULL  - ")

print(rwd_cleaned)

#6th--->>>>:

rwd_mean = rwd.fillna(rwd.mean()['sulphates'])
rwd_median = rwd.fillna(rwd.median()['sulphates'])
rwd_mode = rwd.fillna(rwd.mode()['sulphates'].iloc[0])

#7th--->>>>:

print("null Values with Mean")
print(rwd_mean)

print("null Values with Median")
print(rwd_median)

print("null Values with Mode")
print(rwd_mode)

#8th---->>> Normalizing in one(1) to two(2)
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

#9th--->>>
clm_to_nlz = ['free.sulfur.dioxide', 'total.sulfur.dioxide']

#10th---->>>
rwd_normalized = pd.DataFrame(scaler.fit_transform(rwd[clm_to_nlz]), columns=clm_to_nlz)

#11th--->>>
print("normalized rwd data:")
print(rwd_normalized.head())


#12th--->>> free.sulfur.dioxide and total.sulfur.dioxide

plt.figure(figsize=(7, 5))
plt.subplot(3, 1, 3)
sns.lineplot(x='free.sulfur.dioxide', y='total.sulfur.dioxide', data=rwd)
plt.title('Line Plots for rwd')

#13th---->>>>: for free.sulfur.dioxide and total.sulfur.dioxide
plt.subplot(2, 1, 1)
sns.scatterplot(x='free.sulfur.dioxide', y='total.sulfur.dioxide', data=rwd)
plt.title('Scatter Plots for rwd')

plt.show()

#14th---->>>:
plt.figure(figsize=(7, 3))
sns.boxplot(x='quality',y='alcohol', data=rwd)
plt.title('Boxplots for Wine Quality')

```

#PART 2 : Water Quality

###About Water Quality

####the water_potability.csv file comprises data on water quality metrics for 3276 water bodies. Key parameters include pH values within the recommended WHO range, hardness attributed to calcium and magnesium, total dissolved solids (TDS) affecting taste and color, chloramines as disinfectants, natural sulfate concentrations, electrical conductivity meeting WHO standards, organic carbon levels adhering to EPA guidelines, trihalomethanes in chlorinated water, turbidity indicating suspended solids, and a Potability indicator (1 for safe, 0 for not potable). This dataset offers valuable insights into the safety and quality of water intended for human consumption.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler

#1st--->>>>>>>>>>>
water_quality =pd.read_csv('/content/drive/MyDrive/water_potability.csv')
print(water_quality)

#2nd--->>>>>>>>>>>
print("Indian Water Quality Dataset:")
print(water_quality.head())

#3rd--->>>>>>>>>>>
print("\nNumber of Records - Water Quality:", len(water_quality))

#4th--->>>>>>>>>>>
print("\nNmbr of Clms and Dscrptn - Water Quality:")
print(water_quality.info())

#5th--->>>>>>>>>>>
water_quality_clnd_dropna = water_quality.dropna()

print(water_quality_clnd_dropna)

#6th--->>>>>>>>>>>
water_quality_clnd_mean = water_quality.fillna(water_quality.mean())
print(water_quality_clnd_mean)


water_quality_clnd_median = water_quality.fillna(water_quality.median())
print(water_quality_clnd_median)

water_quality_clnd_mode = water_quality.apply(lambda x: x.fillna(x.mode()[0]))
print(water_quality_clnd_mode)

#7th--->>>>>>>>>>
scaler = MinMaxScaler()
hardness_cdt = ['Hardness', 'Conductivity']
water_quality_nrmlzd = pd.DataFrame(scaler.fit_transform(water_quality[hardness_cdt]), columns=hardness_cdt)
print(water_quality_nrmlzd)

```
